```python
import skimage.io
import numpy as np
import matplotlib.pyplot as plt
import math

def color_distance(c1, c2):
    a = c1[0] - c2[0]
    b = c1[1] - c2[1]
    c = c1[2] - c2[2]
    return math.sqrt(a**2+b**2+c**2)
import functools

# @functools.lru_cache(maxsize=255*255*255)
def get_best_crayola_color(color):    
    mindistance = 255**2
    bestcolor = None
    for crayola in CRAYOLA_COLORS:
        distance = color_distance(color, crayola["rgb_color"]) 
        if distance < mindistance:
            mindistance = distance
            bestcolor = crayola
    
    return bestcolor


image = skimage.io.imread('input_pictures/starrynight.jpg')
image = image.astype(np.float64)
plt.imshow(image/255)
plt.show()

for y in range(image.shape[0]):
    print(y)
    for x in range(image.shape[1]):
        match = get_best_crayola_color(tuple(image[y,x]))
        image[y,x] = match['rgb_color']
```

```python
lookup_table = list()
for x in range(255):
    print(x)
    for y in range(255):
        for z in range(255):
            lookup_table.append(get_best_crayola_color([x,y,z]))
```
